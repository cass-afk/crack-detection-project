```python
import os
import yaml

# Load dataset configuration
data_yaml = 'dataset.yaml'  # Path to your dataset.yaml
with open(data_yaml, 'r') as file:
    data = yaml.safe_load(file)

# Print the dataset paths for debugging
train_path = data['train']
val_path = data['val']
print("Train path:", train_path)
print("Validation path:", val_path)

# Check if the paths exist and list the files
if os.path.exists(train_path):
    print("Training images:", os.listdir(train_path))
else:
    print("Training path does not exist.")

if os.path.exists(val_path):
    print("Validation images:", os.listdir(val_path))
else:
    print("Validation path does not exist.")
```

```python
import matplotlib.pyplot as plt
import yaml
import os
from pathlib import Path

# Training command with hooks for saving and plotting loss
def train_and_plot(img_size, batch_size, epochs, data_yaml, model_cfg, pretrained_weights):
    # YOLOv5 training command
    train_command = f"""python train.py --img {img_size} --batch {batch_size} --epochs {epochs} --data {data_yaml} --cfg {model_cfg} --weights {pretrained_weights}"""

    # Run the training
    print("Starting Training...")
    os.system(train_command)
    print("Training Completed!")

```

```python
train_and_plot(
    img_size=227,
    batch_size=16,
    epochs=50,
    data_yaml="dataset.yaml",
    model_cfg="yolov5s.yaml",
    pretrained_weights="last.pt"
)
```

```python

    # Path to results
results_path = Path("/content/drive/MyDrive/CV_Project/yolov5/runs/train/exp3/results.csv")  # YOLOv5 stores training metrics in CSV
if not results_path.exists():
  raise FileNotFoundError(f"{results_path} not found! Ensure training ran successfully.")

    # Read the results CSV
import pandas as pd
results = pd.read_csv(results_path)

    # Extrac
results.head()
```

```python
columns = results.columns
columns = columns.str.replace(' ','')
results.columns = columns
```

```python
epochs = results['epoch']+101
epochs
```

```python

train_loss = results['train/box_loss'] + results['train/obj_loss'] + results['train/cls_loss']
val_loss = results['val/box_loss'] + results['val/obj_loss'] + results['val_cls/loss']

    # Plot Train and Validation Loss
plt.figure(figsize=(10, 6))
plt.plot(epochs, train_loss, label="Train Loss", color="blue")
plt.plot(epochs, val_loss, label="Validation Loss", color="orange")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Train vs Validation Loss 150 epochs")
plt.legend()
plt.grid()
plt.savefig("train_val_loss_150.png")
print("Saved train vs validation loss plot as 'train_val_loss_150.png'.")

```
